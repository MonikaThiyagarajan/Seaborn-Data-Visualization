# Seaborn Data Visualization 

##### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("tips.csv")
```

### Understanding the dataset

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

###  Distribution plot

```python
sns.distplot(df['tip'])
```

### Distribution Plot with color and bins

```python
sns.distplot(df['total_bill'],color = 'g',bins = 45)
```

### Distribution Plot without kde

```python
sns.distplot(df['total_bill'],bins = 10,kde = False)
```

### Joint Plot

##### Joint Plot with Kind as hex

```python
sns.jointplot(x = 'tip',y = 'total_bill', data = df, kind = 'hex' , color = 'red')
```

##### Joint Plot with Kind as reg

```python
sns.jointplot(x = 'total_bill',y = 'tip', data = df, kind = 'reg',color = 'brown')
```

##### Joint Plot with Kind as kde

```python
sns.jointplot(x = 'total_bill',y = 'tip', data = df, kind = 'kde',color = 'Black')
```

##### Joint Plot with Kind as scatter

```python
sns.jointplot(x = 'total_bill',y = 'tip', data = df, kind = 'scatter',color = 'Orange')
```

### Pairplot

```python
df.head(2)
```

```python
sns.pairplot(df)
```

##### Pair Plot with parameter hue as sex and palette as husl

```python
sns.pairplot(df, hue = 'sex',palette = 'husl')
```

##### Pair Plot with parameter hue as smoker and palette as cubehelix

```python
sns.pairplot(df, hue = 'smoker',palette = 'cubehelix')
```

# Categorical Plots

### Bar Plot

```python
df.head(2)
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = df, palette = 'coolwarm')
```

##### Bar plot with hue as smoker

```python
sns.barplot(x = 'day', y = 'tip', hue = 'smoker', data = df, palette = 'magma')
```

### Count Plot

```python
df['sex'].value_counts()
```

```python
sns.countplot('sex', data=df, palette = 'muted')
```

```python
df['day'].value_counts()
```

```python
sns.countplot('day',data = df,palette = 'hls',hue = 'smoker')
```

### Violin Plots

```python
sns.violinplot(x = 'sex',y = 'total_bill',data = df, palette = 'rocket')
```

```python
sns.violinplot(x = 'day',y = 'total_bill',data = df, palette = 'Accent',hue = 'sex',split= True)
```

### Box Plots

```python
sns.boxplot('day','total_bill',data = df,palette = 'husl')
```

```python
sns.boxplot('day','total_bill',hue = 'time' ,data = df,palette = 'husl')
```

### Strip Plot

```python
sns.stripplot(x = 'day', y = 'tip' , data = df)
```

```python
sns.stripplot(x = 'sex', y = 'tip' , data = df , jitter = False)
```

### Swarm Plot

```python
sns.swarmplot(x = 'day', y = 'tip' , data = df , hue = 'sex')
```

### Factor Plot

```python
sns.factorplot(x = 'day', y = 'total_bill',data = df, kind = 'violin')
```

### Heat Map

```python
df.corr()
```

```python
sns.heatmap(df.corr(), annot = True)
```

```python
sns.heatmap(df.corr(), annot = True, cmap = 'mako', linecolor = 'white' , linewidth = 3)
```

### Cluster Map

```python
sns.clustermap(df.corr() , annot = True)
```
